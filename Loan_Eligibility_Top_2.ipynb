```python

# Define features and target variable
X = dataset.drop(['prediction', 'TempId'], axis=1)  # Drop TempId as it is not a feature
y = dataset['prediction']
X.head()
```

```python
# Z-score normalization
from sklearn.preprocessing import StandardScaler
def normalization (X,y):
    scaler = StandardScaler()
    X_scaled = 0
    X_scaled = scaler.fit_transform(X)

    # Convert the scaled data in x back to a DataFrame
    X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

    # Display the first few rows of the scaled DataFrame
    print("Z-Scored Variables (X): ", + X_scaled_df.head())
    print(X_scaled_df.shape)

    print("\n Target Variable (y):\n", + y.head())
    # Check unique values in the target variable
    unique_classes = y.unique()
    print(f'Unique classes in the target variable: {unique_classes}')
    return X_scaled_df, y
```

```python
#Callling current dataset
X, y = normalization(X, y)
```

```python
X.head()
```

```python
#Method to find top to predictions
def get_top_two_classes(proba):
    top_two = np.argsort(proba, axis=1)[:, -2:][:, ::-1]
    return top_two
```

```python

#Split data into training and testing
from sklearn.model_selection import train_test_split # library

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20) # Specifies the test dataset as 20% of the data, the 80% rest is located to the training

#Ensuring correct data size
print('x_train: ', X_train.shape)
print('y_train: ',y_train.shape)
print('x_test: ', X_test.shape)
print('y_test: ', y_test.shape)
```

```python
# importing reporting
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
###Logistic Regression

#importing library from sklearn
from sklearn.linear_model import LogisticRegression
def logistic_reg (X_test, y_test, X_train, y_train):

    # Initialize and train the Logistic Regression model
    log_reg = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1000, random_state=42)
    log_reg.fit(X_train, y_train)

    # Predict probabilities on the test set
    y_proba_lr = log_reg.predict_proba(X_test)

    # Method to find top two predictions
    def get_top_two_classes(proba):
        top_two = np.argsort(proba, axis=1)[:, -2:][:, ::-1]
        return top_two

    # Get top two predictions
    top_two_predictions_lr = get_top_two_classes(y_proba_lr)

    # Predict probabilities on the test set
    y_proba_lr = log_reg.predict_proba(X_test)

    # Get top two predictions
    top_two_predictions_lr = get_top_two_classes(y_proba_lr)


    print("Logistic Regression Top Two Predictions:")
    for i in range(len(X_test)):
        print(f"Test instance {i}:")
        top_two = top_two_predictions_lr[i]
        if top_two[0] == -1:  # Both classes are marked
            print(f"Top two predicted classes: {top_two}")
            print(f"Probabilities: {y_proba_lr[i][top_two]}")
        else:
            print(f"Top predicted class: {top_two[0]}")
            print(f"Probability: {y_proba_lr[i][top_two[0]]}")
        print()

    # Predict on the test set
    y_pred_lr = log_reg.predict(X_test)

    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred_lr)
    print(f"Accuracy: {accuracy:.2f}")

    # Print classification report
    print(classification_report(y_test, y_pred_lr))

    # Print confusion matrix
    print(confusion_matrix(y_test, y_pred_lr))

    # Predict on the test set
    y_pred_lr = log_reg.predict(X_test)

```

```python
# Call the logistic_reg function
logistic_reg(X_test, y_test, X_train, y_train)
```

```python

### Decision Tree Classifier
#importing library from sklearn
from sklearn.tree import DecisionTreeClassifier

def d_t_c (X_test, y_test, X_train, y_train):

    DT_clf = DecisionTreeClassifier(random_state=42) #callling and instantiating the built-in DecisionTreeClassifier from Sklearn Scikit
    DT_clf.fit(X_train, y_train)#fit method of the class is called to train the algorithm on the training data, which is passed as parameter to the fit method

    y_pred_dt = DT_clf.predict(X_test)

    # Predict probabilities on the test set
    y_proba_dt = DT_clf.predict_proba(X_test)
    # Get top two predictions
    top_two_predictions_dt = get_top_two_classes(y_proba_dt)

    print("Decision Tree Top Two Predictions:")
    for i in range(len(X_test)):
        print(f"Test instance {i}:")
        top_two = top_two_predictions_dt[i]
        if top_two[0] == -1:  # Both classes are marked
            print(f"Top two predicted classes: {top_two}")
            print(f"Probabilities: {y_proba_dt[i][top_two]}")
        else:
            print(f"Top predicted class: {top_two[0]}")
            print(f"Probability: {y_proba_dt[i][top_two[0]]}")
        print()

    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred_dt)
    print(f"Accuracy: {accuracy:.2f}")

    # Print classification report
    print(classification_report(y_test, y_pred_dt))

    # Print confusion matrix
    print(confusion_matrix(y_test, y_pred_dt))

    # Convert the decision tree data back to a DataFrame
    DT_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred_dt})

    DT_df.head()
```

```python

#Neural-Network
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.utils import to_categorical

def nn_model(X_test, y_test, X_train, y_train):

    # Convert labels to categorical
    y_train_categorical = to_categorical(y_train, num_classes=9)
    y_test_categorical = to_categorical(y_test, num_classes=9)

    # Build the neural network model
    model = Sequential()
    model.add(Dense(128, input_dim=X_train.shape[1], activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(128, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(64, activation='relu'))
    model.add(Dense(9, activation='softmax'))

    # Compile the model
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

    # Train the model
    class_weights = {0: 1.0, 1: 1.0, 2: 3.0, 3: 3.0, 4: 3.0, 5: 3.0, 6: 3.0, 7: 3.0, 8: 3.0}
    model.fit(X_train, y_train_categorical, epochs=50, batch_size=32, validation_split=0.2, class_weight=class_weights, callbacks=[tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5)])

    # Predict probabilities on the test set
    y_proba_nn = model.predict(X_test)

    # Get the top two predictions
    top_two_predictions_nn = np.argsort(y_proba_nn, axis=1)[:, -2:][:, ::-1]

    print("Neural Network Top Two Predictions:")
    for i in range(len(X_test)):
        print(f"Test instance {i}:")
        print(f"Top two predicted classes: {top_two_predictions_nn[i]}")
        print(f"Probabilities: {y_proba_nn[i][top_two_predictions_nn[i]]}")

    # Convert probabilities to class predictions
    y_pred_nn = np.argmax(y_proba_nn, axis=1)
    y_test_actual = np.argmax(y_test_categorical, axis=1)

    # Calculate accuracy
    accuracy = accuracy_score(y_test_actual, y_pred_nn)
    print(f"Accuracy: {accuracy:.2f}")

    # Print classification report
    print(classification_report(y_test_actual, y_pred_nn))

    # Print confusion matrix
    print(confusion_matrix(y_test_actual, y_pred_nn))

    return y_proba_nn
```

```python
nn_model(X_test, y_test, X_train, y_train)
```

```python
#Meta-Model
def meta_model(X_test, y_test, X_train, y_train):
    y_proba_nn = nn_model(X_test, y_test, X_train, y_train)
    y_proba_xgb = xgb_model(X_test, y_test, X_train, y_train)

    # Combine the predictions into meta_features
    meta_features = np.hstack((y_proba_nn, y_proba_xgb))

    # Convert y_test back to original label format
    #y_test_labels = np.argmax(y_test, axis=1)

    # Train a meta-model (Logistic Regression)
    meta_model = LogisticRegression(max_iter=1000, random_state=42)
    meta_model.fit(meta_features, y_test)

    # Predict with meta-model on test set
    final_predictions = meta_model.predict(meta_features)

    # Calculate accuracy
    accuracy_meta = accuracy_score(y_test, final_predictions)
    print(f"Meta-Model Accuracy: {accuracy_meta:.2f}")

    # Print classification report
    print(classification_report(y_test, final_predictions))

    # Print confusion matrix
    print(confusion_matrix(y_test, final_predictions))

    for i, prediction in enumerate(final_predictions):
        print(f"Test instance {i}: Predicted class: {prediction}")
```

```python
#Meta model
meta_model(X_test, y_test, X_train, y_train)
```
